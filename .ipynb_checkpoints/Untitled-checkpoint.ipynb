```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

```python
df = pd.read_csv('heart_desease.csv')
df.head()
```

```python
df_cat = df[["HeartDisease", "Smoking", "AlcoholDrinking", "DiffWalking", "Sex", "Race", "Diabetic", "PhysicalActivity", "GenHealth","Asthma", "KidneyDisease", "SkinCancer"]]
df_num = df[['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']]
df_cat
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df_cat.columns:
    le=LabelEncoder()
    label=le.fit_transform(df_cat[i])
    df_cat[i]=label
df_cat.head()
```

```python
data = pd.concat([df_cat, df_num], axis = 1)
data
```

```python
data.corr()
```

```python
data.drop(["BMI", "Race"], axis=1,  inplace =True)
```

```python
data
```

```python
x = data.drop('HeartDisease', axis=1)
y = data.HeartDisease
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
lr = LinearRegression()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
pipeline_log_reg=Pipeline([('log_reg',LogisticRegression(random_state=42))])
pipeline_lr=Pipeline([('lin_reg',LinearRegression())])


pipelines = [pipeline_lr,  pipeline_log_reg]
pipe_dict = {0: 'Linear Regression',1: "Logistic Regression"}
```

```python
pipeline_lr.fit(x_train,y_train)
pipeline_lr.score(x_test, y_test)
lr_pred =  pipeline_lr.predict(x_test)
lr_pred
```

```python
pipeline_log_reg.fit(x_train,y_train)
pipeline_log_reg.score(x_test, y_test)
log_reg_pred =  pipeline_log_reg.predict(x_test)
print(log_reg_pred)
```

```python
log_reg_pred =  pipeline_log_reg.predict(x_test)
for i in log_reg_pred:
    if i == 1:
        print(i)
```

```python
for pipes in pipelines:
    pipes.fit(x_train, y_train)
```

```python

for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model,x_train, y_train, cv=10)
    print(cv_score)
    print("%s: %f" %(pipe_dict[i], cv_score.mean()) )
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, lr_pred)
cm
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, log_reg_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
print("Training Accuracy :", pipeline_lr.score(x_train, y_train))
print("Testing Accuracy :", pipeline_lr.score(x_test, y_test))
```

```python
print("Training Accuracy :", pipeline_log_reg.score(x_train, y_train))
print("Testing Accuracy :", pipeline_log_reg.score(x_test, y_test))
```
